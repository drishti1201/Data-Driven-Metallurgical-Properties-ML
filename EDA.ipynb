```python
# IMPORT LIBRARIES

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. LOAD DATA 

df = pd.read_csv("advanced_metallurgy_dataset.csv")
```

```python
# 2. BASIC DATA OVERVIEW

print("Shape of dataset:", df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)

print("\nFirst 5 rows:")
print(df.head())
```

```python
# 3. MISSING VALUES

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# 4. STATISTICAL SUMMARY

print("\nStatistical Summary:")
print(df.describe())
```

```python
# 5. HISTOGRAMS (Distribution)

df.hist(figsize=(12,10))
plt.suptitle("Feature Distributions")
plt.show()
```

```python
# 6. CORRELATION MATRIX

corr = df.corr()

plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.title("Correlation Matrix")
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.show()
```

```python
# 7. RELATIONSHIPS (IMPORTANT)

# Cooling Rate vs Hardness
plt.figure()
plt.scatter(df["Cooling_Rate"], df["Hardness"])
plt.xlabel("Cooling Rate")
plt.ylabel("Hardness")
plt.title("Cooling Rate vs Hardness")
plt.show()
```

```python
# Grain Size vs UTS
plt.figure()
plt.scatter(df["Grain_Size"], df["UTS"])
plt.xlabel("Grain Size")
plt.ylabel("UTS")
plt.title("Grain Size vs UTS")
plt.show()
```

```python
# 8. PHASE DISTRIBUTION

df[["Ferrite","Pearlite","Martensite"]].plot(kind="box")
plt.title("Phase Distribution")
plt.show()
```

```python
# Skewness
print("Skewness:\n", df.skew())
```
